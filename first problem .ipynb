```python
df=df.dropna(subset=['adv_id'])
```

```python
df['created_at']=pd.to_datetime(df['created_at'])
```

```python
df=df.sort_values(by=['adv_id','created_at'])
df
```

```python
df_group=df.groupby('adv_id')
```

# 1.How long did it take for the user to make their second transaction on average?

```python
def time_to_second_txn(group):
    if len(group) >= 2:
        delta = group.iloc[1]['created_at'] - group.iloc[0]['created_at']
        return delta.total_seconds() / 3600  
    return None
```

```python
second_txn_times = df_group.apply(time_to_second_txn).dropna()
avg_hours_to_second_txn = second_txn_times.mean()
```

```python
print("Average time to second transaction (in hours):", round(avg_hours_to_second_txn, 2))
```

# 2.How many percent of users who did the first transaction also did the second transaction?

```python
users_with_txns = df_group.size()
users_with_second_txn = users_with_txns[users_with_txns >= 2].count()
percent_with_second_txn = (users_with_second_txn / users_with_txns.count()) * 100
```

```python
print("Percentage of users who made a second transaction:", round(percent_with_second_txn, 2), "%")
```

# 3. What was the average transaction amount after the first transaction?
# For example, if a userâ€™s first transaction was 100 INR and they subsequently made
# transactions of 50, 100, and 200 INR, the average would be (50 + 100 + 200) / 3.

```python
def avg_amount_after_first(group):
    if len(group) >= 2:
        return group.iloc[1:]['value_in_paise'].mean() / 100  #It is converting paise to INR
    return None
```

```python
avg_amounts = df_group.apply(avg_amount_after_first).dropna()
avg_txn_amt_after_first = avg_amounts.mean()
```

```python
print("Average transaction amount after the first (in INR): â‚¹", round(avg_txn_amt_after_first, 2))
```

# Assumptions Made:
Only transactions with a valid adv_id (user ID) are analyzed.

All transaction statuses (e.g., PROCESSED, FAILED) are included unless stated otherwise.

The time to the second transaction is calculated as the time difference between the first and second transactions sorted by timestamp, measured in hours.

Users with at least one transaction are considered, and the percentage is based on how many of them made a second transaction.

The average amount after the first transaction includes all subsequent transactions and is converted from paise to INR.
